```python

x_capacity = int(input("Enter Jug 1 capacity:"))
y_capacity = int(input("Enter Jug 2 capacity:"))
end = int(input("Enter target volume:"))
def bfs(start, end, x_capacity, y_capacity):
    path = []
    frontier = []
    frontier.append(start)
    explored = []
    i=1
    #visited.append(start)
    while frontier:
        current = frontier.pop()
        x = current[0]
        y = current[1]
        path.append(current)
        if x == end or y == end:
            print("Goal node reached")
            return path
        if current[0] < x_capacity and ([x_capacity, current[1]] not in explored):
            frontier.append([x_capacity, current[1]])
            explored.append([x_capacity, current[1]])
        if current[0]>0 and ([0,current[1]] not in explored):
            frontier.append([0,current[1]])
            explored.append([0,current[1]])
        if current[1] < y_capacity and ([current[0], y_capacity] not in explored):
            frontier.append([current[0], y_capacity])
            explored.append([current[0], y_capacity])
        if current[1]>0 and ([current[0],0] not in explored):
            frontier.append([current[0],0])
            explored.append([current[0],0])
        if (current[0]+current[1])<=x_capacity and current[1]>0 and([current[0]+current[1],0] not in explored):
            frontier.append([current[0]+current[1],0])
            explored.append([current[0]+current[1],0])
        if (current[0]+current[1])<=y_capacity and current[0]>0 and([0,current[0]+current[1]] not in explored):
            frontier.append([0,current[0]+current[1]])
            explored.append([0,current[0]+current[1]])
        if (current[0]+current[1])>=x_capacity and current[1]>0 and([x_capacity,current[1]-(x_capacity-current[0])] not in explored):
            frontier.append([x_capacity,current[1]-(x_capacity-current[0])])
            explored.append([x_capacity,current[1]-(x_capacity-current[0])])
        if (current[0]+current[1])>=y_capacity and current[0]>0 and ([current[0]-(y_capacity-current[1]),y_capacity] not in explored):
            frontier.append([current[0]-(y_capacity-current[1]),y_capacity])
            explored.append([current[0]-(y_capacity-current[1]),y_capacity])
        print(f"At step {i} ")
        print("Frontier is : \n",frontier)
        print("Explored is : \n",explored)
        i+=1
    return ("Not found")
def gcd(a, b):
    if a == 0:
        return b
    else:
        return gcd(b%a, a)
start = [0, 0]
if end % gcd(x_capacity,y_capacity) == 0:
    print(bfs(start, end, x_capacity, y_capacity))
else:
    print("No solution possible for this combination.")
```
